Berapa total penjualan (nb_sold) dari alpukat dengan jenis organik pada tahun 2017 dalam DataFrame alpukat?

```python
organic2017_sales = avocado[(avocado["type"] == "organic") & (avocado["year"] == 2017)]["nb_sold"].sum()
print("Total Number Sold of Organic Avocados in 2017 is :", organic2017_sales.round(1))
```

Buatlah diagram batang dari total jumlah orang yang tunawisma (homeless)berdasarkan wilayah di dalam DataFrame homelessness. Urutkan bar chart nya secara menurun (descending). Bonus: buat juga diagram batang horizontal.

```python
homelessness_total = homelessness.groupby("region")["total_homeless_people"]\
                    .sum().sort_values(ascending=True)
bar_chart = homelessness_total.plot(kind="bar")

for index, value in enumerate(homelessness_total) :
    bar_chart.text(index, value, int(value), ha="center", va="bottom")

plt.title("Total Number of Homeless People by Region")
plt.xlabel("Region")
plt.ylabel("Total Number of People")

plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
barh_chart = homelessness_total.plot(kind="barh")
for index, value in enumerate(homelessness_total):
    barh_chart.text(value + 12000, index, int(value), ha="center", va="center")

plt.title("Total Number of Homeless People by Region")
plt.xlabel("Total Number of Homeless People")
plt.ylabel("Region")
plt.show()
```

Buat diagram garis dengan dua garis yang mewakili suhu di Toronto dan Roma. Pastikan untuk memberi label plot dengan benar. Bonus: tambahkan legenda untuk kedua garis.

```python
toronto = temperatures[temperatures["city"] == "Toronto"]
rome = temperatures[temperatures["city"] == "Rome"]

toronto_temp = toronto.groupby("date")["avg_temp_c"].mean().plot()
rome_temp = rome.groupby("date")["avg_temp_c"].mean().plot()
plt.title("Average Temperatures in Toronto and Rome")
plt.xlabel("Date")
plt.ylabel("Temperatures (Degree Celcius)")

plt.legend(["Toronto", "Rome"], fontsize="small")
plt.xticks(rotation=45, ha="right")
plt.show()
```
